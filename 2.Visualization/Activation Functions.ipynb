```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def binaryStep(x):
    ''' It returns '0' is the input is less then zero otherwise it returns one '''
    return np.heaviside(x,1)
```

```python
x = np.linspace(-10, 10)
plt.plot(x, binaryStep(x))
plt.axis('tight')
plt.title('Binary Step')
plt.grid(alpha=0.4, ls='-.')
plt.show()
```

```python
def sigmoid(x):
    ''' It returns 1/(1+exp(-x)). where the values lies between zero and one '''

    return 1/(1+np.exp(-x))
```

```python
x = np.linspace(-10, 10)
plt.plot(x, sigmoid(x))
plt.axis('tight')
plt.title('Sigmoid')
plt.grid(alpha=0.4, ls='-.')
plt.show()
```

```python
def RELU(x):
    ''' It returns zero if the input is less than zero otherwise it returns the given input. '''
    x1=[]
    for i in x:
        if i<0:
            x1.append(0)
        else:
            x1.append(i)

    return x1
```

```python
x = np.linspace(-10, 10)
plt.plot(x, RELU(x))
plt.axis('tight')
plt.title('ReLU')
plt.grid(alpha=0.4, ls='-.')
plt.show()
```

```python
def plot(func, yaxis=(-0.5, 2)):
    plt.ylim(yaxis)
    plt.title('ELU')
    plt.grid(alpha=0.4, ls='-.')
    plt.plot(x, func(x))
```

# OBSERVATIONS


 ## 1. GELU (Gaussian Error Linear Unit)
### Performance:
GELU typically demonstrates smooth and stable convergence.
Its non-monotonic nature and adaptive scaling enable it to retain useful information during optimization.
## #Strengths:
Combines ReLU's simplicity with non-linearity similar to Sigmoid.
Excellent performance in modern neural network architectures, especially in transformers.
### Observations:
GELU likely achieves fast convergence and stabilizes effectively over iterations.

## 2. Mish
### Performance:
Mish is a smooth, non-monotonic activation function that provides robust optimization performance.
It has self-regularizing properties, which may reduce the risk of vanishing gradients.
### Strengths:
Offers better generalization compared to ReLU and GELU in certain architectures.
Its smoothness improves gradient flow.
### Observations:
Mish seems to converge slightly slower than GELU but provides better long-term stability.

## 3. ReLU (Rectified Linear Unit)
### Performance:
ReLU is a simple and computationally efficient activation function.
Prone to the "dying ReLU" problem, where neurons can become inactive.
### Strengths:
Extremely fast convergence during the initial training phase.
Well-suited for deep neural networks.
### Observations:
ReLU converges quickly at first but may experience plateaus or instability in some cases.


## 4. SELU (Scaled Exponential Linear Unit)
### Performance:
SELU is designed for self-normalizing neural networks, maintaining mean activation close to zero and standard deviation close to one.
May perform better on tasks requiring consistent normalization.
### Strengths:
Strong convergence properties in architectures requiring normalization.
Reduces vanishing gradients in deep networks.
### Observations:
SELU appears to exhibit steady convergence with a slightly slower initial rate compared to ReLU.


## 5. Sigmoid
### Performance:
Sigmoid suffers from vanishing gradient issues due to its bounded nature.
Limited utility in deep networks but useful for binary classification tasks.
### Strengths:
Effective in scenarios requiring probability-based outputs.
Smooth activation curve.
### Observations:
Sigmoid converges very slowly, likely due to saturation in higher layers.


## 6. Softmax
### Performance:
Softmax is often used in the output layer for multi-class classification tasks.
Not an activation function for intermediate layers, but its behavior in optimization can still be evaluated.
### Strengths:
Provides normalized probabilities for multi-class problems.
### Observations:
Softmax convergence appears slow and unstable compared to GELU, Mish, or ReLU.


## 7. Tanh
### Performance:
Tanh scales the output to the range [-1, 1], improving gradient flow compared to Sigmoid.
Still suffers from vanishing gradient issues for deep networks.
### Strengths:
Useful for tasks requiring output normalization in intermediate layers.
### Observations:
Tanh converges faster than Sigmoid but is slower compared to ReLU or GELU.



## Key Observations from the Images
### Fastest Convergence:
GELU and ReLU show the fastest initial convergence among the seven functions.
Stability:
Mish and SELU demonstrate better long-term stability, avoiding sharp oscillations or stagnation.
### Slow Convergence:
Sigmoid and Softmax converge the slowest, reflecting their limitations in deep learning tasks.
### Balanced Performance:
Tanh strikes a middle ground, with moderate convergence speed and stability.
### Conclusion
Best Overall: GELU and Mish for their smooth convergence, stability, and strong performance in deep networks.
Fastest Convergence: ReLU for initial training speed.
Stable Convergence: SELU for tasks requiring normalization.
Avoid if Possible: Sigmoid and Softmax due to their slow convergence in intermediate layers.
